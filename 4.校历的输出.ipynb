```python
import time
import calendar
k=0
def nowtime():
    t1=time.time()
    localt1=time.localtime(t1)
    localtime=time.asctime(localt1)
    print(localtime)
    print('信阳师范学院校历')
while k<=1:
    k+=1
    time.sleep(1)
    nowtime()
print('2019-2020学年第一学期')
for i in range(1,13):
    print(i,'月',calendar.month(2019,i))
    time.sleep(1)
print('2019-2020学年第二学期')
for j in range(1,13):
        print(j,'月',calendar.month(2020,j))
        time.sleep(1)
```
